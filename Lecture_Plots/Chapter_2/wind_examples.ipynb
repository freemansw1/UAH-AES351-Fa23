```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_different_winds(u_winds, v_winds):
    '''
    Function to plot the same wind field with different styles
    '''
    
    x_sz, y_sz = np.shape(u_winds)

    x_arr = np.arange(0, x_sz, 1)
    y_arr = np.arange(0, y_sz, 1)
    xx, yy = np.meshgrid(x_arr, y_arr)

    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
    ax1.barbs(u_winds, v_winds)
    ax1.set_title("Barbs")
    q = ax2.quiver(u_winds, v_winds)
    ax2.quiverkey(q, 0.95, 1, 10, r'$10 kts$',)
    ax2.set_title("Vectors")
    ax3.streamplot(xx, yy,u_winds, v_winds, color='k')
    ax3.set_title("Streamlines")

    for ax in (ax1, ax2, ax3):
        ax.axis('equal')
        ax.set(xlim=(-1, x_sz), ylim=(-1, y_sz))
```

```python
# pure u winds
size_arr = (7,7)
u_winds = np.full(size_arr,10)
v_winds = np.full(size_arr, 0)
plot_different_winds(u_winds, v_winds)
plt.savefig("east_flow.png")
```

```python
# pure v winds
size_arr =(7,7)
u_winds = np.full(size_arr,0)
v_winds = np.full(size_arr, 10)
plot_different_winds(u_winds, v_winds)
plt.savefig("north_flow.png")
```

```python
#circular flow
size_arr =(7,7)
u_pos = np.arange(0, size_arr[0],1)
v_pos = np.arange(0, size_arr[1],1)
uu, vv = np.meshgrid(u_pos, v_pos)


u_winds = np.full(size_arr,10)
#u_winds = np.cos(vv*np.pi/6)*u_winds
u_winds = -0.5*(vv-3)*u_winds

v_winds = np.full(size_arr, 10)
#v_winds = np.cos(uu*np.pi/6)*v_winds
v_winds = 0.5*(uu-3)*v_winds

plot_different_winds(u_winds, v_winds)
plt.savefig("circular_flow.png")
```
